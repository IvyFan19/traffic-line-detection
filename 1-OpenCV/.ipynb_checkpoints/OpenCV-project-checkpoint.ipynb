# 1. Color Selection
An image contains three colors (R, G, B):
Blue (0, 0, 255); Red(255, 0, 0); Green(0, 255, 0)

In the most of time, trafic lines are usually White (255, 255, 255). 

```python
heigth = image.shape[0]
width = image.shape[1]
print("the shape of the image is: ", image.shape)

# make a copy of this image
output_img = np.copy(image)
```

```python
# set the threshold for three colors, you can change
th_red = 230
th_gre = 230
th_blu = 230

th_rgb = [th_gre, th_gre, th_blu]
thresholds = (image[:,:,0] < th_rgb[0]) | (image[:,:,1] < th_rgb[1]) | (image[:,:,2] < th_rgb[2])
# When pixel value is less than threshold, it will be set as black.
output_img[thresholds] = [0,0,0]
```

```python
def showImagesHorizontally(list_of_files):
    fig = figure(figsize=(20,10))
    number_of_files = len(list_of_files)
    for i in range(number_of_files):
        a=fig.add_subplot(1,number_of_files,i+1)
        image = list_of_files[i]
        imshow(image,cmap='Greys_r')
        axis('off')

imgs_list = [image, output_img]
showImagesHorizontally(imgs_list)
```
